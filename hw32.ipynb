```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("diabetes.csv")
df
```

```python
new_df = df.fillna(df.mean(numeric_only=True))
print(df.isnull().sum())
df.head()
```

```python
columns = ["Glucose","BloodPressure","BMI","Age","Insulin","Outcome"]
hambastegi = df[columns].corr()
print(hambastegi)
```

```python
new =sns.heatmap(hambastegi,annot=True,cmap="coolwarm")
print(new)

```

```python
outcome_hambastegi = hambastegi["Outcome"].drop("Outcome")
max_f = outcome_hambastegi.idxmax()
max_v = outcome_hambastegi.max()
print(f"{max_f},{max_v}")
```
