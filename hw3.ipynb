```python
import numpy as np
days = np.arange(1,31)
cities = ["tehran","mashhad","isfahan","tabriz","shiraz","ahvaz"]
temp_day = np.random.uniform(15,41,size=(30,6))
humidity = np.random.uniform(20,81,size=(30,6))
rainy = np.random.uniform(0,51,size=(30,6))

```

```python
import pandas as pd
cities = ["Tehran", "Mashhad", "Isfahan", "Tabriz", "Shiraz", "Ahvaz"]
data = []
for i, city in enumerate(cities):       
    for day in range(1, 31):           
        data.append([
            city,                       
            day,                    
            temp_day[day-1, i],      
            humidity[day-1, i],    
            rainy[day-1, i]          
        ])
df = pd.DataFrame(data, columns=["City", "Day", "Temperature", "Humidity", "Rainfall"])
df.head(10)
```

```python

mean_values = df.groupby("City")[["Temperature", "Humidity", "Rainfall"]].mean()
print("mean_city")
print(mean_values)
warmest_city = mean_values["Temperature"].idxmax()
coldest_city = mean_values["Temperature"].idxmin()

print(f"\nwarmcity: {warmest_city}")
print(f"cold city:{coldest_city}")
rainy_days = df[df["Rainfall"] > 10].groupby("City").size()
print(rainy_days)
```

```python
import matplotlib.pyplot as plt
isfahan_data = df[df["City"] == "Isfahan"]
plt.scatter(isfahan_data["Temperature"], isfahan_data["Humidity"])
plt.title("isfahan")
plt.xlabel("temp(Â°C)")
plt.ylabel("hum (%)")
plt.show()
```
